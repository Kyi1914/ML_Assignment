# 5. Evaluation

#### Evaluation Method

Evaluation Method:
- MSE       : Mean Square Error
- MAE.      : Mean Absolute Error
- R2 square : root mean square error

## Evaluation for Linear Regression Model

#### predict

```python
y_predict = LR_model.predict(X_test)
```

```python
y_predict
```

#### change dtype of the original Y_test array

```python
type(y_predict)
```

```python
y_test = y_test.values
```

```python
type(y_test)
```

#### MAE

```python
len(y_test)
```

```python
sum = 0
for i,j in zip (y_test, y_predict):
  diff = i-j
  abs_diff = abs(diff)
  sum += abs_diff
  # print ('Absolute diff', abs_diff)

MAE_LR = sum / len(y_test)
```

```python
print(f"MAE of Linear Regression = {MAE_LR:.2f}")
```

#### MSE

```python
sum = 0
for i,j in zip (y_test, y_predict):
  diff = i-j
  sq_diff = diff ** 2
  sum += sq_diff
  # print ('Absolute diff', abs_diff)

MSE_LR = sum / len(y_test)
```

```python
print(f"MSE of Linear Regression = {MSE_LR:.2f}")
```

## Evaluation for Ridge Model

```python
# predict
y_pred = ridge_model.predict(X_test)

# change type to numpy
# y_pred = y_pred.values
type(y_pred)
```

```python
# MAE

sum = 0
for i,j in zip (y_test, y_predict):
  diff = i-j
  abs_diff = abs(diff)
  sum += abs_diff

MAE_R = sum / len(y_test)

print(f"MAE of Ridge Regression = {MAE_R:.2f}")
```

```python
# MSE

sum = 0
for i,j in zip (y_test, y_predict):
  diff = i-j
  sq_diff = diff ** 2
  sum += sq_diff

MSE_R = sum / len(y_test)

print(f"MSE of Ridge Regression = {MSE_R:.2f}")
```
